```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset

dataset = pd.read_csv("CKD.csv")
dataset
```

```python
# Converting the categorical data to numerical data

dataset = pd.get_dummies(dataset, drop_first = True)
dataset
```

```python
# Getting the colum names of the dataset

dataset.columns
```

```python
# Splitting the input and the output

independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent = dataset[['classification_yes']]
dependent
```

```python
# Splitting the train and the test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = None)
```

```python
# Creating the GridSearch Random Forest Model

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid = {'criterion': ['gini', 'entropy', 'log_loss'], 'max_features': ['sqrt', 'log', None]} 

grid_classifier = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 0, n_jobs = 1)
grid_classifier.fit(x_train,y_train)
```

```python
re = grid_classifier.cv_results_
```

```python
y_pred = grid_classifier.predict(x_test)
```

```python
# Printing the confusion Matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("The Confusion Matrix is: \n", cm)
```

```python
# Printing the classification result

from sklearn.metrics import classification_report
c_report = classification_report(y_test, y_pred)
print("The Classification Report: \n", c_report)
```

```python
# Printing the f1_score

from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, y_pred)
print("The F1_Score is :", f1_macro)
```

```python
# Printing the roc_auc_score

from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid_classifier.predict_proba(x_test)[:,1])
```

```python
# Printing the best parameters

print("The best parameters are : {} ".format(grid_classifier.best_params_))
```

```python
# Printing the grid Search Table

table = pd.DataFrame.from_dict(re)
table
```
